```python
import numpy as np
import matplotlib.pyplot as plt
from mterrain.functions import lorentzian_peak
```

**Peak Function**

Peak like behavior is achieved using a scaled Lorentzian function whose graph has a single peak:

$$f\left(x\right)=\frac{h\left(w \big/ 2\right)^2}{\left(x-c\right)^2+\left(w \big/ 2\right)^2}$$

The paramters of the curve dictate the following:

 - $h$ : the height of the peak at its center
 - $c$ : the center of the peak
 - $w$ : the width of the curve at half of $h$ (FWHM)

```python
x = np.linspace(0, 200, 400)
h = [45.0, 35.0, 55.0]
c = [50.0, 85.0, 145.0]
w = [35.0, 22.0, 50.0]

fig, ax = plt.subplots(figsize=(12, 4))

for i in range(3):
    f = lorentzian_peak(h[i], c[i], w[i])
    ax.plot(x, f(x), label=f"h={h[i]}, c={c[i]}, w={w[i]}")
ax.set_xlim(5, 195)
ax.set_aspect('equal')
ax.grid(True)
plt.legend()
plt.show()
```
